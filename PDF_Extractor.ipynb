```python
import os
import re
import pandas as pd
import PyPDF2

# Function to extract text from a PDF file
def extract_text_from_pdf(pdf_file_path):
    text = ""
    try:
        with open(pdf_file_path, 'rb') as pdf_file:
            pdf_reader = PyPDF2.PdfFileReader(pdf_file)
            for page_num in range(pdf_reader.numPages):
                page = pdf_reader.getPage(page_num)
                text += page.extractText()
    except Exception as e:
        print(f"Error extracting text from PDF: {e}")
    return text


# Directory where CVs in PDF format are stored
cv_directory = r'D:/Project/Resume Extractor/data'  # Replace with your CV directory path

# Create an empty list to store extracted details
extracted_details = []

# Loop through CVs
for cv_filename in os.listdir(cv_directory):
    if cv_filename.endswith('.pdf'):
        cv_path = os.path.join(cv_directory, cv_filename)
        cv_text = extract_text_from_pdf(cv_path)

        category = extract_category(cv_text)
        skills = extract_skills(cv_text)
        education = extract_education(cv_text)

        extracted_details.append({
            'CV_File': cv_filename,
            'Category': category,
            'Skills': ', '.join(skills),
            'Education': ', '.join(education)
        })

# Create a DataFrame from the extracted details
extracted_details_df = pd.DataFrame(extracted_details)

# Save the extracted details to a CSV file
extracted_details_df.to_csv('extracted_cv_details.csv', index=True)

print("Extraction completed.")
```

```python
import numpy as np
from transformers import DistilBertTokenizer, DistilBertModel
import torch
import re

# Function to preprocess text
def preprocess_text(text):
    # Add your preprocessing logic here (e.g., removing special characters)
    # Example: Remove non-alphanumeric characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text

# Modified embed_text function
def embed_text(text):
    # Preprocess the text
    text = preprocess_text(text)
    
    # Tokenize the preprocessed text
    input_ids = tokenizer.encode(text, add_special_tokens=True, max_length=512, truncation=True)
    input_ids = torch.tensor(input_ids).unsqueeze(0)  # Batch size of 1
    
    # Embed the tokenized text
    with torch.no_grad():
        embeddings = model(input_ids)[0]
    
    return embeddings


# Tokenize and embed text
def embed_text(text):
    input_ids = tokenizer.encode(text, add_special_tokens=True, max_length=512, truncation=True)
    input_ids = torch.tensor(input_ids).unsqueeze(0)  # Batch size of 1
    with torch.no_grad():
        embeddings = model(input_ids)[0]
    return embeddings

# Calculate cosine similarity
def calculate_similarity(embedding1, embedding2):
    # Reshape the embeddings for correct dot product
    embedding1 = embedding1.reshape(1, -1)
    embedding2 = embedding2.reshape(1, -1)
    
    similarity = np.dot(embedding1, embedding2.T) / (np.linalg.norm(embedding1) * np.linalg.norm(embedding2))
    return similarity

# Example: Match and rank CVs for multiple job descriptions
job_descriptions = ["Job description 1", "Job description 2", ...]  # Replace with actual job descriptions
cvs = ["CV 1 text", "CV 2 text", ...]  # Replace with actual CV text
top_matching_cvs = []

for job_description in job_descriptions:
    job_description_embedding = embed_text(job_description)
    cv_scores = []

for cv_text in cvs:
    cv_embedding = embed_text(cv_text)
    similarity_score = calculate_similarity(job_description_embedding, cv_embedding)
    cv_scores.append((cv_text, similarity_score))

    # Sort CVs by similarity score in descending order
    cv_scores.sort(key=lambda x: x[1], reverse=True)

    # Store the top 5 matching CVs
    top_matching_cvs.append({
        'Job_Description': job_description,
        'Top_Matching_CVs': cv_scores[:5]
    })

# Display the top matching CVs for each job description
for result in top_matching_cvs:
    print(result['job_Description'])
    for cv, score in result['Top_Matching_CVs']:
        print(f"CV: {cv}, Similarity Score: {score}")
```
